```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vc_Data = np.load("total_visibilities_viscpu_256.npy")
pvs_Data = np.load("total_visibilities_pyuvsim_256.npy")
blpairs = list(np.load("baselines.npy"))
freqs = np.load('freqs.npy')
```

```python
vc_Data.shape
```

```python
total_bls_list = [(x, y, 'xx') for (x, y) in blpairs]
total_waterfall_labels = ['Baseline = {}'.format(bl) for bl in total_bls_list]
```

```python
n=772
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].plot(freqs, np.real(vc_Data[n][0,:]),label='VisCPU')
ax[0].plot(freqs, np.real(pvs_Data[n][0,:]),label='pyuvsim')
ax[1].plot(freqs, np.abs((np.real(vc_Data[n][0,:]) - np.real(pvs_Data[n][0,:])) * 100. / np.real(pvs_Data[n][0,:])))
ax[0].legend()
fig.suptitle(total_waterfall_labels[n])
ax[0].set_title('Re(V) vs Frequency')
ax[1].set_title('Percentage error in Re(V) vs Frequency')
```

```python
fig2, ax2 = plt.subplots(1, 2, figsize=(12, 6))
ax2[0].plot(freqs, np.imag(vc_Data[n][0,:]),label='VisCPU')
ax2[0].plot(freqs, np.imag(pvs_Data[n][0,:]),label='pyuvsim')
ax2[1].plot(freqs, np.abs((np.imag(vc_Data[n][0,:]) - np.imag(pvs_Data[n][0,:])) * 100. / np.imag(pvs_Data[n][0,:])))
ax2[0].legend()
fig2.suptitle(total_waterfall_labels[n])
ax2[0].set_title('Im(V) vs Frequency')
ax2[1].set_title('Percentage error in Im(V) vs Frequency')
```
